## Looking at the underlying trend, there appears to be a linear upward trending relationship between Advertising Budget and Sales. Sales tends to be higher the  larger the Advertising Budget. Therfore, we may be able to develop a linear model to predict sales based on advertising budget.

```python
df_sales_advrts_sorted.corr('pearson')
```

Correlation Coefficient is a measure of how values of one variable (Sales) moves with values of another variable (Advertising Budget).

the value of correlation coeeficient ranges from -1.00 to +1.00, with 1.00 representing a prefect linear relationship, where, for example, Advertising Budget
values can be used to perfectly predict Sales. In visual speak, every value of Advertising and associated sales falls on a straight line. There is no scatter
around the trend line. The negative sign means there is an inverse relationship (meaning as Advertising Budget increases, sales tend to fall). Poitive sign
imply positive relationship - as Advertising Budget increases, so does the sales. 

A value closer to 1.00 is desirable. In our case, the Correlation Coeeficient = 0.631428. strengthwise, this indicates a weak to medium linear relationship between 
Advertising Budget and Sales. Further, it is positive as the Corr. Coeff. is a positive value. Therefore, the data suggests a positive, linear relationship between Sales and advertising budget. Knowledge of advertising budget may help predict sales.  

Overall, it is worth attempting to fit a linear model (of the form y = mx + b) to the data.
